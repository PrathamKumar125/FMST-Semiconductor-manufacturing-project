# Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
scaler = StandardScaler().fit(df2.drop("Pass/Fail" , axis=1))
```

```python
X= scaler.transform(df2.drop("Pass/Fail" , axis=1))
Y= df2["Pass/Fail"]
X_train , X_test ,Y_train, Y_test=train_test_split(X,Y,test_size=0.10 , random_state=0)
```

# Classification Model

```python
from sklearn.metrics import confusion_matrix , accuracy_score
```

```python
acc=[]
model_name=['LR','DT','PER','RF','KNN']
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression(multi_class='multinomial', solver='saga', max_iter=10000)
model.fit(X_train, Y_train.values.ravel())
```

```python
y_pred=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
acc_lr=accuracy_score(Y_test,y_pred)
acc.append(acc_lr)
acc_lr
```

```python
print(confusion_matrix(Y_test,y_pred))
```

# DecisionTreeClassifier

```python
from sklearn.tree import DecisionTreeClassifier
clf=DecisionTreeClassifier(criterion='gini',max_depth=1000)
clf.fit(X_train, Y_train.values.ravel())
```

```python
y_pred=clf.predict(X_test)
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
acc_decisionT=accuracy_score(Y_test,y_pred)
acc.append(acc_decisionT)
acc_decisionT
```

# Perceptron

```python
from sklearn.linear_model import Perceptron
clf3=Perceptron()
clf3.fit(X_train, Y_train.values.ravel())
```

```python
y_pred=clf3.predict(X_test)
conf_mat=confusion_matrix(Y_test,y_pred)
```

```python
sns.heatmap(conf_mat,annot=True,cbar=False,cmap='summer')
```

```python
acc_per=accuracy_score(Y_test,y_pred)
acc.append(acc_per)
acc_per
```

# RandomForestClassifier

```python
from sklearn.ensemble import RandomForestClassifier
cl_random=RandomForestClassifier(n_estimators=20,criterion='gini',max_depth=1000)
cl_random.fit(X_train, Y_train.values.ravel())
```

```python
y_pred=cl_random.predict(X_test)
print(confusion_matrix(Y_test,y_pred))
```

```python
acc_rfc=accuracy_score(Y_test,y_pred)
acc.append(acc_rfc)
acc_rfc
```

# K Nearest Neighbors

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(X_train,Y_train)
y_pred=knn.predict(X_test)
```

```python
print(confusion_matrix(Y_test,y_pred))
```

```python
acc_knn=accuracy_score(Y_test,y_pred)
acc.append(acc_knn)
acc_knn
```

```python
print(acc)
```
