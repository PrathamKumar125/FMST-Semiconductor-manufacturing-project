```python
df1 = df.replace(np.NaN, 0)
```

```python
df2= df1.drop('Time',axis=1)
df2
```

```python
df2.info()
```

```python
df2['Pass/Fail'] = df2['Pass/Fail'].replace(to_replace=[1,-1],value=[0,1])
df2
```

```python
labels = ['Pass', 'Fail']
size = df2['Pass/Fail'].value_counts()
colors = [ 'green', 'orange']
explode = [0.05, 0]
plt.figure(figsize=(8,7))
plt.pie(size, labels =labels, colors = colors, explode = explode, autopct = "%.2f%%", shadow = True)
plt.suptitle("Number of features in dataset", y=0.9, size=18,weight="bold")
plt.legend()
plt.show()
```
